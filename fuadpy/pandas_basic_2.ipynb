# create plots in pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("air_quality_no2.csv")
```

```python
air_quality.head(10)
```

 #  quick visual check of the data.


```python
air_quality.plot()
plt.show()
```

```python
#plot only one data table
air_quality["station_antwerp"].plot()
#plt.show()
```

```python
#visually compare two column of a datafram
air_quality.plot.scatter(x = "station_london" , y = "station_paris" , alpha=0.5)
```

```python
#box function in pandas
air_quality.plot.box()
```

```python
# the columns in a separate subplot.
axs = air_quality.plot.area(figsize=(12 , 4) , subplots=True)
plt.show()
```

```python
# further customize, extend or save the resulting plot.
fig , axs = plt.subplots(figsize=(12,4))
air_quality.plot.area(ax = axs)
axs.set_ylabel("NO$_2$ concentration")
fig.savefig("No2_concentrations.png")
plt.show()
```

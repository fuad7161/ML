# Subset of a dataframe

```python
#Load data
age_sex = titanic[["Age" , "Sex"]]
age_sex.head(10)
```

```python
age_sex.shape
```

# Collect data with some certain condition

```python
above_35 = titanic[titanic["Age"] > 35]
above_35.head(10)
```

```python
#multiple condition fulfilled
class_23 = titanic[(titanic["Pclass"].isin([2,3])) & (titanic["Age"] > 35)]
class_23.head(10)
```

```python
#Remove NaN value from dataFrame
age_no_Nan = titanic[titanic["Age"].notna() & (titanic["Age"] > 35)]
age_no_Nan.head(10)
```

# Find all specific name of with the required condition
## the names of the passengers older than 35 years

```python
adult_names = titanic.loc[titanic["Age"]>35,"Name"]
adult_names.head(10)
```
